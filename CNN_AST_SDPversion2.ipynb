```python
import re
import pandas as pd
import en_core_web_lg

import numpy as np

from sklearn.model_selection import train_test_split


from tensorflow.keras import datasets, layers, models
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('camel1.2AST.csv',header=None)
```

```python
data
```

```python
data=data.dropna()
X=data.iloc[:,:-1]
X=X.values
y=data.iloc[:,-1]
```

```python
len(X[0])
```

```python
nlp = en_core_web_lg.load()
tap_train=[]
for i in range(595):
    e=[]
    for j in range(1111):
        b=[]
        bb=[]
        if X[i][j]==0 :
            for a in range(300):
                bb.append(0)
            b.append(bb)
        else:
            c= re.findall('[A-Z][^A-Z]*', X[i][j])
            d=''
            for k in c:
                d += k
            b.append(nlp(d).vector)
        e.append(b)
    tap_train.append(e)
tap_train=np.asarray(tap_train)
# print(tap_train.shape)
# print(tap_train)
aaa=[]
for dd in range(595):
    for ee in range(1111):
        for ff in range(1):
            for ii in range(300):
                aaa.append(tap_train[dd][ee][ff][ii])
tap_train=np.asarray(aaa)

tap_train=tap_train.reshape(-1,1111,300,1)
# encoder_y = LabelEncoder()
# y = encoder_y.fit_transform(y)
a = 0
b = 0
status = []
for i in y:
    if i > 0:
        status.append(1)
        a = a + 1
    else:
        status.append(0)
        b = b + 1
y = np.array(status)
# kf = KFold(n_splits=10)
# kf.get_n_splits(y)
# acc=0
# precision=0
# recall=0
# f1=0

X_train, X_test, y_train, y_test = train_test_split(tap_train, y, test_size=0.1, random_state=42,stratify=y)
model = models.Sequential()
model.add(layers.Conv2D(2, (30, 30), activation='relu', input_shape=(1111, 300, 1)))
model.add(layers.Conv2D(4, (50, 50), activation='relu'))
model.add(layers.Flatten())
model.add(layers.Dense(20,activation='relu'))
model.add(layers.Dense(1,activation='sigmoid'))
model.compile(loss='BinaryCrossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()
history = model.fit(X_train, y_train, epochs=100)
a=model.predict(X_test)
y_pred=[]
for c in range(len(a)):
  if a[c] < 0.5:
    y_pred.append(0)
  else:
    y_pred.append(1)
from sklearn.metrics import accuracy_score, f1_score, recall_score, precision_score
print(accuracy_score(y_test,y_pred))
print(f1_score(y_test,y_pred))
print(recall_score(y_test,y_pred))
print(precision_score(y_test,y_pred))
```
